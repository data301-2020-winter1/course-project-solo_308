```python
from Scripts import project_functions as pf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pf.load_and_process('https://raw.githubusercontent.com/data301-2020-winter1/course-project-solo_308/main/vgsales.csv?token=ARDWNGUU74PCHQDADL2ZC627WYEGS')
df
```

```python
# I am trying to find which platform has perfromed better worldwide
df2 = df.groupby(['Platform']).mean()
df2 = df2.sort_values('Global_Sales', ascending = False)
df2 = df2.reset_index(col_level = 1)
df2
```

```python
fig, ax = plt.subplots(figsize=(8,12))
sns.barplot(x='Global_Sales', y= 'Platform', data = df)
plt.xlabel('Global_Sales')
# Game boy(GB) has been the best perfroming platfrom globally in terms of game sales.
```

```python
fig, ax = plt.subplots(figsize=(8,12))
sns.barplot(x='NA_Sales', y= 'Platform', data = df)
plt.xlabel('NA_Sales')
# Nintendo Entertainment System(NES) has been the best perfroming platfrom in North America in terms of game sales.
```

```python
fig, ax = plt.subplots(figsize=(8,12))
sns.barplot(x='EU_Sales', y= 'Platform', data = df)
plt.xlabel('EU_Sales')
# Game Boy(GB) has been the best perfroming platfrom in Europe in terms of game sales.
```

```python
fig, ax = plt.subplots(figsize=(8,12))
sns.barplot(x='JP_Sales', y= 'Platform', data = df)
plt.xlabel('JP_Sales')
# Nintendo Entertainment System(NES) has been the best perfroming platfrom in Japan in terms of game sales.
```
